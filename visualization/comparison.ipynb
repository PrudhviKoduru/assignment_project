```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your simulation results
df = pd.read_csv("rc_results.csv")

plt.figure(figsize=(9,6))

# Overlaid distributions
sns.kdeplot(data=df[df["path"]=="edge"], x="total_delay_ms", label="Edge", fill=True, alpha=0.4)
sns.kdeplot(data=df[df["path"]=="rc"], x="total_delay_ms", label="Regional", fill=True, alpha=0.4)
sns.kdeplot(data=df[df["path"]=="cloud"], x="total_delay_ms", label="Cloud", fill=True, alpha=0.4)

plt.title("Delay Distribution Comparison: Edge vs Regional vs Cloud")
plt.xlabel("Total Delay (ms)")
plt.ylabel("Density")
plt.legend(title="Processing Layer")
plt.xlim(0, df["total_delay_ms"].quantile(0.95))  # trim extreme outliers
plt.tight_layout()
plt.show()

```
