```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("rc_results.csv")

# Average delay by path
df.groupby("path")["total_delay_ms"].mean().plot(kind="bar", title="Average Delay by Layer")
plt.ylabel("Delay (ms)")
plt.show()

# Cost distribution
df.groupby("path")["cost_usd"].mean().plot(kind="bar", color="orange", title="Cost by Path")
plt.ylabel("USD")
plt.show()

# Scatter for energy vs delay
df.plot.scatter(x="total_delay_ms", y="energy_j", alpha=0.3, c=df["path"].astype("category").cat.codes, colormap="viridis")
plt.title("Energy vs Delay across Layers")
plt.show()
```
